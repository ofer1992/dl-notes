This is just a test of viability.  
The parameter estimation problem is, having the N-point data set ${x[0],x[1],\dots,x[N-1]}$ which depends on an unknown parameter $\theta$, define an estimator
$$\hat{\theta}=g(x[0],x[1],\dots,x[N-1])$$

that determines $\theta$.

## Example: Dow-Jones industrial average
Let's conjecture the dow jones to be a line embedded in random noise,
$$x[n] = A + Bn +\omega[n]\qquad n=0,1,\dots,N-1$$
So our parameters are $\theta=[AB]^{T}$, and we'll assume $\omega[n] \equiv \cal{N}(0,\sigma^2)$.

I'll skip writing the pdf and let's try to see how the data would look like.

```python
# Let's get the actual data for comparison
dja = quandl.get('BCB/UDJIAD1', start_date='2014-03-07', end_date='2014-07-29')
n = dja.size
```

```python
# parameters and data generation
A = 16200
B = 10
sigma = 120
w = np.random.normal(0, sigma, size=n)
x = A + B*np.arange(n) + w
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(dja.size), dja)
plt.plot(x)
plt.legend(['dja', 'synthetic'])
```

So after some "manual" estimation I arrived in some parameters which generate synthetic dow jones data that doesn't seem too far from the real one.
