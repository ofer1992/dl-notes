The triads built on scale degrees:

```python
ex1 = stream.Stream()
k = key.Key("C")
for i in range(1, 8):
    c = k.romanNumeral(i)
    c.lyric = c.romanNumeralAlone
    c.quarterLength = 1
    ex1.append(c)
ex1.show()
```

The real meaning of Roman numerals is the triads and their _harmonic function_. The relationship between _root chords_ is called _root relationships_ , and a series of two or more where the root relationship is an important factor is called a _root progression_.

The voice compasses:

```python
voices_treb = stream.Stream()
voices_bass = stream.Stream()
notes = ['b3', 'c4', 'a5', 'b5', 'e3', 'f3', 'c5', 'd5']
for n in notes[:8]:
    voices_treb.append(note.Note(n))
notes = ['c3', 'c3', 'g4', 'a-4', 'c2', 'd2', 'c4', 'd4']
for n in notes[:8]:
    voices_bass.append(note.Note(n))
voices_bass.clef= clef.BassClef()
print('soprano                       alt')
voices_treb.show()
print('tenor                         bass')
voices_bass.show()
```
