```python
key = s.analyze('key')
```

```python
chord.Chord([key.getTonic(), key.get, 0])
```

```python
pitches = key.getPitches()
tonic = chord.Chord([pitches[i] for i in [0, 2, 4]])           
dominant = chord.Chord([pitches[i] for i in [4, 6, 1]])           
subdominant = chord.Chord([pitches[i] for i in [3, 5, 0]])           
m = stream.Measure(number=1)
m.repeatAppend(tonic, 2)
m.append(subdominant)
m.append(dominant)
m.append(chord.Chord([pitches[i] for i in [0, 2, 4]]))
m.show()
m.show('midi')
```
